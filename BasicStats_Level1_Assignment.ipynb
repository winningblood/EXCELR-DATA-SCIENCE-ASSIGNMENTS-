# Standardization

Transforming the whole data in between -3 to +3 and removing the units from the data is called Standardization.

Then tha data lies b/w -3 to +3:

1)99.7% of data lies always b/w -3 to +3 sigma approximately.

2)95.5% of data lies always b/w -2 to +2 sigma approximately.

3)68% of data lies always b/w -1 to +1 sigma approximately.

Standard Normal Value:

Z=x-Î¼/Ïƒ;

Z-Standardized value(Zscore); 

x-original value of data point; 

Î¼-mean of the dataset; 

Ïƒ-standard deviation of the dataset;

This is how the zscore will be calculated of all the numerical columns in the dataset:

```python
from statsmodels.stats import weightstats as ztests
```

```python
zcalc,pvalue=ztests.ztest(df['Volume'],value=5,alternative = 'smaller')
```

```python
print('zcalc value:',zcalc)
```

# Importance of Data Preprocessing:

Standardizatoin:
Standardization using z-score normalization for ensuring that numerical features contribute equally to the analysis or modeling process.By standardizing, you ensure that all features are on a similar scale, which improves the performance and accuracy of models.

One Hot Encoding:
One-hot encoding is essential for converting categorical variables into a numerical format that machine learning models can process. Without encoding, models that require numerical input wouldnâ€™t be able to interpret categorical variables.
However, care must be taken to avoid introducing multicollinearity

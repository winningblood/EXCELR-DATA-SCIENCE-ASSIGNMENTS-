# Dimensionality Reduction with PCA:

```python
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA
```

```python
scaler = StandardScaler()# standardize the data(mean =0,std =1)
data_scaled = scaler.fit_transform(df)
```

```python
pca = PCA()
pca.fit(data_scaled)
```

```python
#Scree Plot
plt.figure(figsize=(10,5))
plt.plot(range(1,len(df.columns)+1),pca.explained_variance_ratio_,marker ='o')
plt.xlabel("Principal Component")
plt.ylabel("Explained Variance Ratio")
plt.title("Scree Plot")
plt.show()
```

```python
import numpy as np
#cumulative explained variance plot
cum_var = np.cumsum(pca.explained_variance_ratio_)
plt.figure(figsize =(10,5))
plt.plot(range(1,len(cum_var)+1),cum_var,marker="o")
plt.xlabel('Number of Components')
plt.ylabel('Cumulative Explained Variance')
plt.title('Cumulative Explained Variance Plot')
plt.show()
```

```python
#optimal number of components
optimal_components = np.argmax(cum_var >= 0.95) + 1
print("Optimal Number of Components:",{optimal_components})
```

```python
#transform data into principal components
pca = PCA(n_components = optimal_components)
data_pca = pca.fit_transform(data_scaled)
print("Shape of PCA transformed Data:",data_pca.shape)
```

# Clustering with Original Data:

```python
from sklearn.cluster import KMeans
from sklearn.metrics import silhouette_score
#Apply Kmeans
kmeans_orig = KMeans(n_clusters = 3,random_state = 42)
kmeans_orig.fit(data_scaled)
labels_orig = kmeans_orig.labels_
```

```python
plt.scatter(data_scaled[:,0],data_pca[:,1],c=labels_orig,cmap = 'viridis')
plt.title("K-means Clustering on Original Data(Visualized using PCA)")
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```

```python
#silhouette score of original data
sil_score_orig = silhouette_score(data_scaled,labels_orig)
print("Silhouette Score(Original data):",sil_score_orig.round(3))
```

# Clustering with PCA:

```python
kmeans_pca = KMeans(n_clusters = 3 , random_state =42)
kmeans_pca.fit(data_pca)
labels_pca = kmeans_pca.labels_
```

```python
plt.scatter(data_pca[:,0],data_pca[:,1],c=labels_pca,cmap='viridis')
plt.title('K-Means Clustering on PCA transformed data')
plt.xlabel('PCA1')
plt.ylabel('PCA2')
plt.show()
```

```python
#silhouette score
sil_score_pca = silhouette_score(data_pca,labels_pca)
print("Silhouette score (PCA data):",sil_score_pca.round(3))
```

# Comparision:

```python
#silhouette scores:
print("Silhouette Score(Original data):",sil_score_orig.round(3))
print("Silhouette score (PCA data):",sil_score_pca.round(3))

#differences in cluster labels
cluster_diff = np.sum(labels_orig != labels_pca)
print("Number of Different lables:",cluster_diff)
```
